```python
MotifResults = namedtuple('MotifResult', 'motif pvalue bonf_pvalue counts')
```

```python
def decode_motif_result(motif, pvalue, bonf_pvalue, counts):
    return MotifResults(motif, float(pvalue), float(bonf_pvalue), counts)


def get_rank(results, target_motif):
    results.sort(key=lambda r: r.bonf_pvalue)
    
    for rank, motif_result in enumerate(results):
        if motif_result.motif == target_motif:
            print(motif_result)
            return rank
    return None

for condition in target_conditions:
    path = os.path.join(scratch_dir, condition, 'motif.case-control.tsv')
    results = ! cat {path}
    results = results[1:]
    results = [decode_motif_result(*row.split()) for row in results]
    
    target_motif = motif_for_condition[condition]
    rank = get_rank(results, target_motif)
    
    print(condition, target_motif, rank)
```
